```scala
/**
  * Transforms [[io.eels.Row]] onto type A
  */
// Ideally, we would like to employ a view, but this will have to wait for Scala 2.13 or even Dotty.
// https://scala-lang.org/blog/2017/11/28/view-based-collections.html
import scala.util.Try
trait ParquetTransformer[A] extends Iterable[Try[A]] {
  def transform: (io.eels.Row) => Try[A]
}
```

```scala
/**
  * Transform a [[io.eels.component.parquet.ParquetSource]] onto an [[Iterator[A]]
  */
trait ParquetSourceTransformer[A] extends ParquetTransformer[A] {
  import io.eels.component.parquet.ParquetSource
  val parquet: ParquetSource
}
```

```scala
trait DisplayTransformer extends ParquetSourceTransformer[String] {
  import io.eels.Row
  import org.json4s._
  import org.json4s.native.JsonMethods._
  import scala.util.{Success,Failure}

  override def transform: Row => Try[String] = parseRow
  override def iterator: Iterator[Try[String]] = parquet.toDataStream.iterator.map(row => transform(row))

  private def parseRow(row: Row): Try[String] = {
    //FIXME: a code review is needed here: it's not clear if a merge will destroy some useful information.
    val slices = row.values.map(row => row.toString).filter(row => row.startsWith("{"))
    val json = slices.map(text => parse(text)).reduce(_ merge _)
      
    val gatewaySerial  = (json \\ "gateway" \\ "serialNumber") .values
    val gatewayArticle = (json \\ "gateway" \\ "articleNumber").values
    val deviceSerial   = (json \\ "device"  \\ "serialNumber") .values
    val deviceArticle  = (json \\ "device"  \\ "articleNumber").values
      
    val text = s"${gatewaySerial}*${gatewayArticle}*${deviceSerial}*${deviceArticle}"
    Success(json.toString)
  }
}
object DisplayTransformer {
  import io.eels.component.parquet.ParquetSource
  import org.apache.hadoop
  class wrapper(val parquet: ParquetSource) extends DisplayTransformer
  def apply(parquet: ParquetSource): DisplayTransformer = new wrapper(parquet)
  def apply(file: java.io.File): DisplayTransformer = apply(file.toURI)
  def apply(url: java.net.URL): DisplayTransformer = apply(url.toURI)  
  def apply(uri: java.net.URI): DisplayTransformer = {
    implicit val hadoopConfiguration = new hadoop.conf.Configuration
    implicit val hadoopFileSystem = hadoop.fs.FileSystem.get(hadoopConfiguration)
    new wrapper(ParquetSource(uri))
  }
  def apply(source: String): DisplayTransformer = {
    implicit val hadoopConfiguration = new hadoop.conf.Configuration
    implicit val hadoopFileSystem = hadoop.fs.FileSystem.get(hadoopConfiguration)
    val parquet = ParquetSource(source)
    new wrapper(parquet)
  }
}
```
